```python
# A list of my grades

my_grades = ["B", "C", "B", "D"]

# convert my grades into a series 

my_grades = pd.Series(my_grades)
my_grades
```

```python
# map - map("current_value1" : "new_value1", "current_value2" : "new_value2")MY

my_grades.map({"B": "A", "C" : "B", "D": "C"})
```

```python
# format function

my_grades = [92.34, 84.56, 86.78, 98.32]

for grade in my_grades:
    print("{:.1f}".format(grade))
```

```python
train = pd.read_csv("http://bit.ly/kaggletrain")
train.head()
```

```python
# using .map() | allows you to map a new value to an existing value in a SERIES

train["Sex_num"] = train.Sex.map({"female": 0, "male": 1})
```

```python
train.loc[0:5, ["Sex", "Sex_num"]]
```

```python
train["Name_length"] = train.Name.apply(len)
```

```python
train.loc[0:5, ["Name", "Name_length"]]
```

```python
import numpy as np
```

```python
train["Fare_ceil"] = train["Fare"].apply(np.ceil)
```

```python
train.loc[0:5, ["Fare", "Fare_ceil"]]
```

```python
train["Name"].str.split(",").head()

# commas are the seperators in a list for the elements
```

```python
def get_element(my_list, position):
    return my_list[position]
    
```

```python
train["Name"].str.split(",").apply(get_element, position=0).head()

# the position=0 is just the index, which we defined in the function above 0=1st element, 1 = second element
```

```python
# lambda function version of the above 

train["Name"].str.split(",").apply(lambda x: x[0]).head()
```
